### Exercici 3

Creem una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
import numpy as np   # Importem la biblioteca numpy

# Creem una matriu amb nombres enters de dimensió 5x5

arrayB = np.random.randint(10, size=(5,5))
print(arrayB)

max_value =np.max(arrayB)
print("\nEl valor màxim del array és: ", max_value)

max_row = np.max(arrayB, axis=0)
print("El valor màxim del l'eix de les files és: ", max_row)
                                                       
max_column = np.max(arrayB, axis=1)
print("El valor màxim del l'eix de les columnes és: ", max_column)                                                 
```

### Exercici 4

Mostrem amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
import numpy as np

# Broadcastable arrays:  Numpy descriu com tractar les matrius amb diferentes formes en les operacions aritmètiques. 
# Per determinar si dos matrius son compatibles les seves dimensions han de coincidir o una de les dimensions té una mida d'1.

print("4-Dimensional Arrays with the same dimension\n")

arrayC = np.random.randint(10, size=(4,6))
print("ArrayC: 4x6 \n", arrayC)

arrayD = np.random.randint(10, size=(4,6))
print("\nArrayD: 4x6 \n", arrayD)

print("\n ArrayC x ArrayD\n", (arrayC * arrayD))
```

```python
print("Scalar and 8-Dimensional Array\n")

arrayE = np.random.randint(10, size=(8,12))
print("ArrayE: 8x12 \n", arrayE)

c = 4
print("\nScalar: 4\n", c)

print("\nArrayE + 4 \n", (arrayE + c))
```

```python
print("3-Dimensional and 1-Dimensional Arrays\n")

arrayF = np.random.randint(10, size=(3,7))
print("ArrayF: 3x7 \n", arrayF)

arrayG = np.random.randint(10, size=(3))
print("\nArrayD: 3x \n", arrayG)

print("\n ArrayF - ArrayG\n", (arrayC - arrayD))
```

```python
print("1-Dimensional and 3-Dimensional Arrays\n")

arrayH = np.random.randint(10, size=(3))
print("ArrayH: 3x \n", arrayH)

arrayJ = np.random.randint(10, size=(3,1))
print("\nArrayD: x3 \n", arrayJ)

print("\n ArrayH / ArrayJ\n", (arrayH / arrayJ))
```

### Exercici 5

Utilitzem la Indexació per extreure els valors d'una columna i una fila de la matriu. I sumem els seus valors.

```python
import numpy as np

arrayK = np.random.randint(10, size=(8,8))
print("\nArrayK: 8x8 \n", arrayK)

row3 = arrayK[2, :]
print("\nExtracció de la fila 3 del array\n", row3)

column4 = arrayK[: , 3]
print("\nExtracció de la columna 4 del array\n", column4)

print("\nSuma dels valors estrets\n", (row3 * column4))
```

### Exercici 6

Mask la matriu anterior, realitzem un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre.

Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

```python
# 

import numpy as np

arrayM = np.random.randint(0, 50, size=[5,7])
print("\nArrayM: 5x7 \n", arrayM)

mask_array = (arrayM % 4 == 0)
print("\nArrayM mask: 5x7 \n", mask_array)
```

```python
# Fem una copia a modificar
img2 = img.copy()

# Eliminem el canal G Verd 
img2[:,:,0] = 0

img_show = plt.imshow(img2)
```
